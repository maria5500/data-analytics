**Exercici 1. Comptador i endreçador de paraules d'un text.**

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

```python
# Importar Python String Module
import string

def word_occurrence(file_path):
    # Amb un diccionari comptarem les aparicions de cada paraula
    diccionari = {}

    # Obrir l'arxiu, llegir i retornar tot el text
    with open(file_path, 'r', encoding='utf-8') as arxiu:
        text = arxiu.read()
    
    # Dividir el text en paraules 
    words = text.split()

    # Diccionari per comptar les aparicions de cada paraula
    occurrence = {}

    # Crear el traductor per eliminar els símbols de puntuació
    translator = str.maketrans('', '', string.punctuation)

    # Iterar sobre cada paraula
    for word in words:
        # Eliminar els símbols de puntuació de la paraula
        clean_word = word.translate(translator).lower()
        # Assegurar-se que la paraula neta no estigui buida i només contingui lletres
        if clean_word.isalpha():
            # Incrementar el comptador de la paraula neta en el diccionari
            occurrence[clean_word] = occurrence.get(clean_word, 0) + 1

    # Organitzar les paraules segons la lletra inicial
    for word, count in occurrence.items():
        key = word[0]
        # Assegurar-se que la clau és una lletra
        if key.isalpha():
            if key not in diccionari:
                diccionari[key] = {}
            diccionari[key][word] = count

    # Ordenar el diccionari per les claus
    diccionari_ordenat = {k: dict(sorted(v.items())) for k, v in sorted(diccionari.items())}

    return diccionari_ordenat

# Ruta de l'arxiu
file_path = "tu_me_quieres_blanca.txt"

# Cridar la funció word_occurrence amb el text de l'arxiu
resultat = word_occurrence(file_path)

# Mostrar el resultat
print("Aquest és el resultat:")
for letter, words in resultat.items():
    print(f"{letter}:")
    for word, count in words.items():
        print(f"  {word}: {count}")
```
